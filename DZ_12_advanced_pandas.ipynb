### Задание 1

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:
* если источник traffic_source равен yandex или google, то в source_type ставится organic
* для источников paid и email из России - ставим ad
* для источников paid и email не из России - ставим other
* все остальные варианты берем из traffic_source без изменений

```python
import pandas as pd
```

```python
df = pd.read_csv('C:\\Users\\HP\\Desktop\\DataAnalitics\\Python\\Notebook\\доп_материалы_к_пандас_1_и_2\\visit_log.csv', sep=';')
```

```python
df.head()
```

```python
# используем фильтр с методом loc, выбирая столбцы в соответствии с условиями

df.loc[(df.traffic_source == 'yandex') | (df.traffic_source == 'google'), 'source_type'] = 'organic'
df.loc[(df.traffic_source == 'paid') & (df.region == 'Russia') | (df.traffic_source == 'email') & (df.region == 'Russia'), 'source_type'] = 'ad'
df.loc[(df.traffic_source == 'paid') & (df.region != 'Russia') | (df.traffic_source == 'email') & (df.region != 'Russia'), 'source_type'] = 'other'

# используем фильтр на пустые значения столбца
df.loc[(pd.isnull(df.source_type)), 'source_type'] = df['traffic_source']
```

### Задание 3

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
df_rat = pd.read_csv('C:\\Users\\HP\\Desktop\\DataAnalitics\\Python\\Notebook\\ml-latest-small_02\\ratings.csv', sep=',')
```

```python
df_rat.head()
```

```python
# сгруппируем по userId и посчитаем кол-во оценок и мин/макс дату
df_userid = df_rat.groupby('userId').agg({'rating': 'count', 'timestamp': ['min', 'max']}).reset_index()[['userId', 'rating', 'timestamp']]
df_userid.head()
```

```python
# добавляем новый столбец, и создадим новый ДФ с пользователями, которые выставили более 100 оценок
df_userid['rating_100'] = df_userid['rating'] > 100
df_filtered = df_userid.loc[df_userid.rating_100 == True]
df_filtered.head()
```

```python
# и посчитаем для них среднее время жизни в днях
average_lifetime = round((df_filtered['timestamp', 'max'] - df_filtered['timestamp', 'min']).mean() / 86400)
average_lifetime
```

### Задание 4

Дана статистика услуг перевозок клиентов компании по типам (см. файл “Python_13_join.ipynb” в разделе Материалы для лекции «Продвинутый pandas» ---- Ноутбуки к лекции «Продвинутый pandas»).

Необходимо сформировать две таблицы:
* таблицу с тремя типами выручки для каждого client_id без указания адреса клиента
* аналогичную таблицу по типам выручки с указанием адреса клиента

Обратите внимание, что в процессе объединения таблиц данные не должны теряться.

Дана статистика услуг перевозок клиентов компании по типам:

* rzd - железнодорожные перевозки
* auto - автомобильные перевозки
* air - воздушные перевозки
* client_base - адреса клиентов

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
rzd_and_auto = rzd.merge(auto, how = 'outer', on = 'client_id')
rzd_and_auto
```

```python
all = rzd_and_auto.merge(air, how = 'outer', on = 'client_id').fillna(0)
all
```

```python
purchases_with_address = client_base.merge(all, how = 'outer', on = 'client_id')
purchases_with_address
```

```python
purchases_without_address = client_base.merge(all, how = 'outer', on = 'client_id')
purchases_without_address.loc[:, ['client_id', 'rzd_revenue', 'auto_revenue', 'air_revenue']].head(10)
```
